```python
import pandas as pd
import ast
```

```python
df = pd.read_excel("evaluasi.xlsx")
df
```

```python
df["retrieved"] = df["retrieved"].apply(ast.literal_eval)
df["relevant"] = df["relevant"].apply(ast.literal_eval)
df
```

```python
df["intersection"] = df.apply(lambda row: list(set(row['retrieved']) & set(row['relevant'])), axis=1)
df
```

```python
df["true_positive"] = df["intersection"].apply(len)
df["false_positive"] = df.apply(lambda row: len(row["retrieved"]) - row["true_positive"], axis=1)
df
```

```python
tp = df["true_positive"].sum()
fp = df["false_positive"].sum()
precision = tp / (tp + fp)
print(f"Precision: {precision}")
```
